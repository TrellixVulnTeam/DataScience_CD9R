# Regressão Linear
` Definição`

"Dada uma coleção de dados amostrais emparelhados, a equação de regressão
$$\displaystyle \hat{y} = b_{0} + b_{1} x$$
descreve algebricamente a relação entre as duas variáveis $x$ e $y$. O Gráfico da 
equação da regressão é chamada de reta de regressão(ou reta do melhor ajuste, ou reta de mínimos quadrados.)</br>
Podemos perceber que o $y$ é uma variavel dependente do $x$, portando $y$ é denominado 
sendo nossa variavél dependente. Já o $x$ é a nossa váriavel explanátoria ou também chamada
de variável independente.
Essa primeira equação que foi demonstrada é referênte a uma equação de regressão
linear simples, podemos ter regressões lineares múltiplas, a formula abaixo é uma representação de uma equação multipla:
$$\displaystyle \hat{y} = b_{0} + b_{1} x_{1}+b_{2}x_{2} + ... + b_{k} x_{k}$$ 
Agora temos as váriavéis explatórias sendo $(x_{1}, x_{1},... ,x_{k})$ e continuamos tendo nossa váriavel dependente sendo $y$
